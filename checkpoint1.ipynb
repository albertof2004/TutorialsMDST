Let's compare Michigan to Ohio. Which state has the higher average across all years in the test you chose?

```python
def highest_avg(df):
    michigan_score = df.loc[df['STATE'] == "MICHIGAN", 'AVG_MATH_8_SCORE'].mean()
    ohio_score = df.loc[df['STATE'] == "OHIO", 'AVG_MATH_8_SCORE'].mean()
    if michigan_score > ohio_score:
        return "Michigan"
    else:
        return "Ohio"
print (highest_avg(df))
```

<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable. 

You can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

Use this space to modify or create features.

```python
df['TOTAL_EXPENDITURE_PER_STUDENT'] = df['TOTAL_EXPENDITURE'] / df['ENROLL_ALL']

df['TOTAL_INSTRUCTIONAL_EXPENDITURE_PER_STUDENT'] = df['INSTRUCTION_EXPENDITURE'] / df['ENROLL_ALL']

bins = [0, 1000000, 5000000, 10000000, 50000000, np.inf]
labels = ['Low', 'Medium', 'High', 'Very High', 'Ultra High']
df['TOTAL_REVENUE_BIN'] = pd.cut(df['TOTAL_REVENUE'], bins=bins, labels=labels)
```

Feature engineering justification: 
1) and 2): As mentioned in the tutorial, states have different amounts of people. Analyzing the total expenditure and the total instructional expenditure per student could give us more precise insights. 
3):Binning reduces the significance of small differences in data and objectively groups total revenue into broader categories

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = df[['AVG_READING_8_SCORE', 'AVG_READING_4_SCORE', 'AVG_MATH_4_SCORE']].dropna()
y = df.loc[X.index]['AVG_MATH_8_SCORE']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=0)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
print("model intercept:", model.intercept_)
print("model coef:", model.coef_)

prediction_errors = model.predict(X_test) - y_test
mean_prediction_error = np.mean(prediction_errors)
print("mean prediction error:", mean_prediction_error)

accuracy = model.score(X_test, y_test)
print("model accuracy:", accuracy)
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = 'AVG_READING_8_SCORE'

f = plt.figure(figsize=(12, 6))
plt.scatter(X_train[col_name], y_train, color="red", label='True Training')
plt.scatter(X_train[col_name], model.predict(X_train), color="green", label='Predicted Training')

plt.legend()
plt.xlabel(col_name)
plt.ylabel('Math 8 score')
plt.title("Model Behavior On Training Set")
plt.show()
```

```python
col_name = 'AVG_READING_8_SCORE'

f = plt.figure(figsize=(12, 6))
plt.scatter(X_test[col_name], y_test, color="blue", label='True testing')
plt.scatter(X_test[col_name], model.predict(X_test), color="black", label='Predicted testing')

plt.legend()
plt.xlabel(col_name)
plt.ylabel('Math 8 score')
plt.title("Model Behavior on Testing Set")
plt.show()
```
